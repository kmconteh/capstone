```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./datasets/all indicators_final_clean.csv')

pd.set_option('max_columns', 300)
df.head(2)
```

The below outlines the correlation between select data sets with the target y-variable (INFORM epidemic risk score) to identify the optimal x variables for predictive modeling.

### Overall correlation of indicators to INFORM risk score

```python
#dropping non-country indicator columns
df_corr = df.drop(columns = ['Country Name', 'Country Code', 'Region', 'Income Group', 
                             'Sub-Region', 'INFORM: Hazard & Exposure risk score', 'INFORM: Vulnerability risk score', 
                             'INFORM: Lack of Coping Capacity risk score', 'INFORM: Region Hazard & Exposure risk score', 
                             'INFORM: Region Vulnerability risk score', 'INFORM: Region Lack of Coping Capacity risk score', 
                             'INFORM: Region Epidemic risk score', 'INFORM: Income Group Epidemic risk score',
                             'INFORM: Income Group Hazard & Exposure risk score', 'INFORM: Income Group Vulnerability risk score', 
                             'INFORM: Income Group Lack of Coping Capacity risk score'])
```

```python
# created heatmap for initial overview of all country indicators correlation to INFORM epidemic risk score
plt.figure(figsize=(8, 12))
sns.heatmap(df_corr.corr()[['INFORM: Epidemic risk score']].sort_values('INFORM: Epidemic risk score'),
            annot=True,
            cmap='coolwarm',
            vmin=-1,
            vmax=1);
```

Based on the above correlations to the target y-variable, the following were selected as the final indicators to use for the model:
- Death by communicable diseases (% of total) 
- Employment in agriculture, female (%)
- Employment in agriculture, male (%) 
- Employment in industry, female (%)
- Employment in services, female (%) 
- Employment in services, male (%) 
- Fixed broadband subscriptions (# per 100 people) 
- Financing to implement IHR (score out of 100) 
- Medical doctors (# per 10,000) 
- Nursing and midwife personnel (# per 10,000)
- Population ages 15-64 (% of total population) 
- Wage and salaried workers, female (%) 
- Wage and salaried workers, male (%) 
- Handwashing with soap (%)
- Population ages 0-14 (% of total population)
